```python
import cv2
import mediapipe as mp
import time

cap=cv2.VideoCapture(0)
pTime=0

mpDraw=mp.solutions.drawing_utils
mpFaceMesh=mp.solutions.face_mesh
faceMesh=mpFaceMesh.FaceMesh(max_num_faces=2)  #max_num_faces = 2 means its can detect two faces at the time 
drawSpec=mpDraw.DrawingSpec(thickness=1,circle_radius=2)

while True:
    success,img=cap.read()
    if not success:
        break
    imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=faceMesh.process(imgRGB)
    if result.multi_face_landmarks:
        for face_lms in result.multi_face_landmarks:
           # mpDraw.draw_landmarks(img,face_lms,mpFaceMesh.FACEMESH_TESSELATIO,drawSpec,drawSpec)
            mpDraw.draw_landmarks(img, face_lms, mpFaceMesh.FACEMESH_TESSELATION, drawSpec, drawSpec)
            for id,lm in enumerate(face_lms.landmark):    #467 landmarks 
                #print(lm)
                ih,iw,ic=img.shape
                x,y=int(lm.x*iw),int(lm.y*ih)     #to find the extact position of pixel
                print(id,x,y)

       

    cTime=time.time()
    fps=1/(cTime-pTime)
    pTime=cTime
    cv2.putText(img,f'FPS:{int(fps)}',(20,70),cv2.FONT_HERSHEY_PLAIN,2,(45,255,56),3)
    cv2.imshow("YOU",img)
    if cv2.waitKey(1) & 0xff == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

           Use FACEMESH_TESSELATION  If you need a full face structure (AR, 3D animation).
           Use FACEMESH_CONTOURS  If you only need important facial features (face shape, filters).
           Use FACEMESH_IRISES  If you want to track eye movements (gaze tracking, eye contact detection)
